#### Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df2 = application_df.drop(columns=['EIN','NAME'], axis=1)
```

```python
# Determine the number of unique values in each column.
application_df2.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df2['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = []

for i in range(len(application_df2['APPLICATION_TYPE'].value_counts())):
    if application_df2['APPLICATION_TYPE'].value_counts()[i] < 156:
        application_types_to_replace.append(application_df2['APPLICATION_TYPE'].value_counts().index[i])


# Replace in dataframe
for app in application_types_to_replace:
    application_df2['APPLICATION_TYPE'] = application_df2['APPLICATION_TYPE'].replace(app,"Other")


# Check to make sure binning was successful
application_df2['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df2['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df2['CLASSIFICATION'].value_counts().loc[lambda x: x>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced

# use the variable name `classifications_to_replace`
classifications_to_replace = []

for i in range(len(application_df2['CLASSIFICATION'].value_counts())):
    if application_df2['CLASSIFICATION'].value_counts()[i] < 30:
        classifications_to_replace.append(application_df2['CLASSIFICATION'].value_counts().index[i])
```

```python
# Replace in dataframe

for cls in classifications_to_replace:
    application_df2['CLASSIFICATION'] = application_df2['CLASSIFICATION'].replace(cls,"Other")  

    
    # Check to make sure binning was successful
application_df2['CLASSIFICATION'].value_counts()
```

```python
# Generate our categorical variable lists
charity_cat = application_df2.dtypes[application_df2.dtypes == "object"].index.tolist()
```

```python
charity_cat
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
converted_app = pd.get_dummies(application_df2, columns=charity_cat, drop_first=True)
```

```python
converted_app
```

```python
# Split our preprocessed data into our features and target arrays
y = converted_app['IS_SUCCESSFUL']
X = converted_app.drop('IS_SUCCESSFUL', axis=1)
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2022, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled.shape[1]
```

## Use Keras-Tuner to find best parameters to optimize the model

```python
# Create a method that creates a new Sequential model with hyperparameter options:  method based off of class activities
def create_model(hp):
    nn_model = tf.keras.models.Sequential()

    # Allow keras-tuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','selu'])
    
    # Allow keras-tuner to decide number of neurons in first layer
    # Input dimensions set to X_train_scaled.shape[1] to be automatically set to the number of features(columns)
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=80,
        step=5), activation=activation, input_dim=X_train_scaled.shape[1]))

    # Allow keras-tuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=80,
            step=5),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    
    return nn_model
```

```python
# Import the keras-tuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    overwrite=True,
    objective="val_accuracy",
    max_epochs=100,
    hyperband_iterations=10)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=50,validation_data=(X_test_scaled,y_test))
```

```python
# Get the top 2 models. from https://keras.io/guides/keras_tuner/getting_started/
models = tuner.get_best_models(num_models=2)
best_model = models[0]
# Build the model.
# Needed for `Sequential` without specified `input_shape`.
best_model.build(input_shape=(None, 28, 28))
best_model.summary()
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_inputs = X_train.shape[1]
hidden_layer1 = 66
hidden_layer2 = 36
hidden_layer3 = 21
hidden_layer4 = 11
hidden_layer5 = 76

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer1, input_dim=number_inputs, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer3, activation="relu"))

# Fourth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer4, activation="relu"))

# Fifth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_layer5, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adamax", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
